```python
# Filtering out the warnings

import warnings
warnings.filterwarnings('ignore')

# Importing the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# LOADING THE DATASET
df_ss = pd.read_csv('SampleSuperstore.csv')
```

```python
# CHECKING FOR THE STATISTICAL INFORMATION USING THE DESCRIBE FUNCTION
df_ss.describe()
```

```python
# MAKING A LIST OF NAMES OF ALL THE COLUMNS WHOSE DATATYPE IS OBJECT
list_categorical = [feature for feature in df_ss.columns if df_ss[feature].dtype == 'O']
list_categorical
```

```python
# MAKING A LIST OF NAMES OF ALL THE NUMERICAL COLUMNS
list_numerical = [feature for feature in df_ss.columns if df_ss[feature].dtype != 'O']
list_numerical
```

```python
# FINDING THE CORRELATION BETWEEN THE NUMERICAL COLUMNS
df_ss.corr()
```

```python
# DROPPING THE COLUMN POSTAL CODE AS IT DOES NOT HELP IN OUR ANALYSIS.
df_ss = df_ss.drop(['Postal Code'], axis = 1)
```

```python
# PLOTTING A HEATMAP 
plt.figure(figsize=(10,10))
sns.heatmap(df_ss.corr(), annot=True)
```
